```python
#загузка используемых библиотек
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
from scipy import stats

%matplotlib inline
```

```python
sns.set(
    font_scale =1.25,
    style      ="whitegrid",
    rc         ={'figure.figsize':(10,5)}
)
```

```python
#сгененируем распределение
mu = 100
sigma = 10
population = np.round(np.random.normal(mu, sigma, 1000), 2)
```

```python
sns.distplot(population, kde = True)
```

```python
n = 1000
sample_size_z = 100
sample_means_z = []
for i in range(n):
    sample = np.random.choice(population, sample_size_z, False)
    sample_means_z.append(sample.mean())
    
    
n = 1000
sample_size_t = 5
sample_means_t = []
for i in range(n):
    sample = np.random.choice(population, sample_size_t, False)
    sample_means_t.append(sample.mean())
```

## First Visualization

```python
np.std(sample_means_t) / np.sqrt(sample_size_t)
```

```python
# Визулализируем распределение среднее значение выборки по 100 наблюдениям
sns.distplot(sample_means_z, kde = True)
```

```python
# Визулализируем распределение среднее значение выборки по 5 наблюдениям
sns.distplot(sample_means_t, kde = True)
```

```python
# делаем стандартизацию

sample_z = [(x-np.mean(sample_means_z)) / np.std(sample_means_z) for x in sample_means_z]

sample_t = [(x-np.mean(sample_means_t)) / np.std(sample_means_t) for x in sample_means_t]
```

```python
population_z = [(x-np.mean(population)) / np.std(population) for x in population]

n = 1000

sample_size_z = 100
sample_means_z = []
for i in range(n):
    sample = np.random.choice(population_z, sample_size_z, False)
    sample_means_z.append(sample.mean())

sample_size_t = 30
sample_means_t = []
for i in range(n):
    sample = np.random.choice(population_z, sample_size_t, False)
    sample_means_t.append(sample.mean())
```

```python
sns.distplot(sample_means_z, kde = True)
```

```python
sns.distplot(sample_means_t, kde = True)
```
